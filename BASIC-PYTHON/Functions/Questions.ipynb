```python
def make_multiplier(n):
    def multiplier(x):
        return x*n
    return multiplier
double = make_multiplier((2))
print(double(5))
```

```python
f = lambda x: x+3
print(f(2))
```

```python
add = lambda a, b: a+b
print(add('3','4'))
```

```python
square = lambda x: x*x
print(square(5))
```

```python
nums = [1, 2, 3]
res = list(map(lambda x: x * 2, nums))
print(res)
```

```python
nums = [1, 2, 3, 4]
evens = list(filter(lambda x: x%2 == 0, nums))
print(evens)
```
